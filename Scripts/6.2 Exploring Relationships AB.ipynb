```python
#importing merged data for chi-squared 
df_merged = pd.read_csv(os.path.join(path, 'Data', 'data_merged.csv'))
```

```python
import pandas as pd
import scipy.stats as stats

# Step 1: Create a contingency table
contingency_table = pd.crosstab(df_merged['Target'], df['Course'])

# Step 2: Run the chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Step 3: Print results
print("Chi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("p-value:", p)

# Display the expected frequencies
expected_df = pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns)
print("\nExpected Frequencies:")
display(expected_df)
```

### Hypothesis and Questions Explored
Question 1: What student characteristics are most predictive of dropout during 1st year?

Question 2: Are there specific courses where dropout rates are higher? 

Question 3: How does a studenst academic performance in highschool or their entry pathway affect their likelihood of graduating on time?

Question 4: Can we segment students into clusters based on dropout likelihood and academic success and tailor interventions accrodingly?

Question 5:  How do financial variables influence dropout and academic performance?

Hypothesis: 
1. There is higher female enrollment than male but the dropout rates are similar for both genders. Students who dropout tend to score lopwer on both the 1st and 2nd semester grades. 
2. The courses with the highest dropout rates are Management (both daytime and evening programs). While Nursing has the highest number of graduates, it also experiences a dropout rate comparable to that of Management.
3. Those with higher previous qualification (grade) and students with scholarships have higher drop out rates. 
4. Admission grade, Previous qulaification grade and course all can be used to create segments for dropout and academic interventions.
5. The tuition fees being up to date, debtor, and displaced students do not seem to impact dropout rates as much. The varibles that do, however, are scholarship holder. Students tend to score better on average in both semesters when they have a scholarship. Financail varibles do not seem to influence academic performance.
